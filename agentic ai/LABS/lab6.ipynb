```python
lab-6:
 Predict if a student passes or fails based on study hours – your first ML model
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np


X = np.array([
    [1],
    [2],
    [3],
    [4],
    [5],
    [6],
    [7],
    [8]
])


y = np.array([
    0,
    0,
    0,
    1,
    1,
    1,
    1,
    1
])


model = LogisticRegression()
model.fit(X, y)

result = model.predict([[5]])

print("Prediction:", result[0])

if result[0] == 1:
    print("Pass")
else:
    print("Fail")
```

```python
hours = float(input("Enter study hours: "))

result = model.predict([[hours]])

if result[0] == 1:
    print("Pass")
else:
    print("Fail")
```
